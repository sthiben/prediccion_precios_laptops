```python
from sklearn.model_selection import  train_test_split
```

```python
x = df_ordinal_encoder.drop(columns='Price_euros') # Todas las varibales menos la variable objetivo
y = df_ordinal_encoder['Price_euros'] # solo la variable objetivo

x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(x,
                                                    y.values.reshape(-1,1),
                                                    train_size = 0.8,
                                                    random_state = 23102024,
                                                    shuffle = True)
```

```python
# Creacion del modelo
import statsmodels.api as sm
x_train_1 = sm.add_constant(x_train_1, prepend=True)
modelo = sm.OLS(endog = y_train_1,
                exog = x_train_1)
modelo = modelo.fit()
print(modelo.summary())
```

```python
# supuestos linealidad
# Test Ramsey
import statsmodels.stats.diagnostic as smd
test_ramsey = smd.linear_reset(modelo, power=2)

# Buscamos un p valor mayor al 0.05 para decir que el modelo es lineal
print('p valor:',test_ramsey.pvalue)
```

```python
# Supuesto independencia de los errores
from statsmodels.stats.stattools import durbin_watson
dw = durbin_watson(modelo.resid)
# se busca un valor entre 1.5 y 2.5 pÃ¡ra que pase el supuesto
print('valor durbin watson:',dw)
```

```python
# supuesto de homocedasticidad
breush_pagan = sm.stats.diagnostic.het_breuschpagan(modelo.resid, x_train_1)
# se busca un p valor menor o igual al 0.05 para que pase el supuesto
breush_pagan[1] # p valor
#p_valor = round(breush_pagan[1],5)
#p_valor
```

```python
x_test_1.head(1)
```

```python
# Predicciones
x_test = sm.add_constant(x_test_1, prepend= True)
predicciones_mod_1 = modelo.get_prediction(exog = x_test).summary_frame(alpha = 0.05)
predicciones_mod_1
predicciones_mod_1['mean'] = round(predicciones_mod_1['mean'],1)
predicciones_mod_1['mean']
```

```python
from sklearn.metrics import mean_squared_error
error_modelo = np.sqrt(mean_squared_error(y_true = y_test_1,
                                          y_pred = predicciones_mod_1['mean']))


print('error del modelo:', round(error_modelo,1))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import RandomizedSearchCV
import multiprocessing
from sklearn.metrics import mean_squared_error, r2_score
```

```python
x_train_3, x_test_3, y_train_3, y_test_3 = train_test_split(
                                  df_ordinal_encoder.drop(columns=['Price_euros']),
                                  df_ordinal_encoder['Price_euros'],
                                  train_size = 0.75,
                                  random_state = 6112204,
                                  shuffle= True)# mesclar datos
```

```python
parametros = {'hidden_layer_sizes': [(10), (10,10), (8, 8)],
              'alpha': np.logspace(-3,3,3),
              'learning_rate_init': [0.001, 0.1]}
```

```python
modelo_regresion = RandomizedSearchCV(
    estimator = MLPRegressor(solver = 'lbfgs', max_iter=1000),
    param_distributions = parametros,
    n_iter= 50,
    scoring = 'neg_mean_squared_error',
    n_jobs = multiprocessing.cpu_count()-1,
    cv = 3,
    verbose = 0,
    random_state = 61212024,
    return_train_score = True,
    error_score='raise'
)

modelo_regresion.fit(X = x_train_3, y = y_train_3)
resultados = pd.DataFrame(modelo_regresion.cv_results_)
```

### Cuarto Modelo

```python
from sklearn.model_selection import train_test_split
```

```python
price_euros = df['Price_euros'].copy()
laptop_data = df.drop(['Price_euros'], axis=1)
```

### Split dataset

```python
# Calculate test and validation set size:
original_count = len(laptop_data)
training_size = 0.60 # 60% of records
test_size = (1 - training_size) / 2


training_count = int(original_count * training_size)
test_count = int(original_count * test_size)
validation_count = original_count - training_count - test_count

print(training_count, test_count, validation_count, original_count)
```

```python
train_x_4, rest_x_4, train_y_4, rest_y_4 = train_test_split(laptop_data, price_euros, train_size=training_count)
test_x_4, validate_x, test_y_4, validate_y = train_test_split(rest_x_4, rest_y_4, train_size=test_count)

print(len(train_x_4), len(test_x_4), len(validate_x))
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
```

```python
one_hot_encoder.fit(train_x_4[['Company','Product', 'TypeName', 'OS', 'Screen',
                    'Touchscreen', 'IPSpanel', 'RetinaDisplay', 'CPU_company',
                    'CPU_model', 'PrimaryStorageType', 'SecondaryStorageType',
                    'GPU_company', 'GPU_model']])
one_hot_encoder.transform(train_x_4[['Company','Product', 'TypeName', 'OS', 'Screen',
                    'Touchscreen', 'IPSpanel', 'RetinaDisplay', 'CPU_company',
                    'CPU_model', 'PrimaryStorageType', 'SecondaryStorageType',
                    'GPU_company', 'GPU_model']])
```

###Pipeline de TransformaciÃ³n

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import FeatureUnion, Pipeline
```

```python
one_hot_encoding = ColumnTransformer([
    (
        'one_hot_encode',
        OneHotEncoder(sparse_output=False, handle_unknown="ignore"),
        [
            'Company','Product', 'TypeName', 'OS', 'Screen',
                    'Touchscreen', 'IPSpanel', 'RetinaDisplay', 'CPU_company',
                    'CPU_model', 'PrimaryStorageType', 'SecondaryStorageType',
                    'GPU_company', 'GPU_model'
        ]
    )
])
```

```python
feature_engineering_pipeline = pipe = Pipeline(
    [
        (
            "features",
            FeatureUnion(
                [
                    ("categorical", one_hot_encoding)
                ]
            ),
        )
    ]
)
```

```python
transformed = feature_engineering_pipeline.fit_transform(train_x_4)
transformed.shape
```

```python
transformed
```

```python

# Get a fresh copy of the pipeline
from sklearn.base import clone

feature_transformer = clone(feature_engineering_pipeline)

features_train_x = feature_transformer.fit_transform(train_x_4)
features_validate_x = feature_transformer.transform(validate_x)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import LinearSVC

model = RandomForestRegressor(n_estimators=100)

model.fit(features_train_x, train_y_4)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
pred_y_4 = model.predict(features_validate_x)
mse = mean_squared_error(validate_y, pred_y_4)
rmse = mse**0.5  # O usar np.sqrt(mse)
r2 = r2_score(validate_y, pred_y_4)
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"RÂ²: {r2}")
```

###ConstrucciÃ³n del pipeline final

```python
final_inference_pipeline = Pipeline([
    ("feature_engineering", clone(feature_engineering_pipeline)),
    ("model", RandomForestRegressor(n_estimators=100))
])
```

```python
final_training_dataset = pd.concat([train_x_4, validate_x])
final_training_response = pd.concat([train_y_4, validate_y])
```

```python
final_inference_pipeline.fit(final_training_dataset, final_training_response)
```
